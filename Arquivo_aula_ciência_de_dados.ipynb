```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
policia = pd.read_csv("police.csv")
policia
```

```python
policia.info()
```

```python
policia.describe()
```

```python
policia.isnull().sum()
```

```python
# Checar correlação
# Calcular a correlação de Pearson
correlacao_pearson = policia.corr(method='pearson')

# Calcular a correlação de Spearman
correlacao_spearman = policia.corr(method='spearman')

# Imprimir a correlação de Pearson
print("Correlação de Pearson:")
correlacao_pearson
```

```python
# Imprimir a correlação de Spearman
print("\nCorrelação de Spearman:")
correlacao_spearman
```

```python
policia['driver_age_raw'] = policia['driver_age_raw'].values.astype(np.int64)
policia['driver_age'] = policia['driver_age'].values.astype(np.int64)
policia.head(50)
```

```python
d=['county_name',"search_type"]
policia=policia.drop(columns=d)
policia
```

```python
media = policia["driver_age"].mean()
policia["driver_age"] = policia["driver_age"].fillna(media)
display(policia)
```

```python
tabela = policia.dropna()
tabela
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
sns.heatmap(tabela.corr(), annot=True, cmap = 'Reds')
```
